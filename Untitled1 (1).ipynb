```python
# ОММ
df = a1
def aver_val(df, n):
  return np.sum(df)/n

df_sq = np.power(df, 2)
def sample_var(df):
  ans = aver_val(df_sq, len(df)) - (aver_val(df, len(df)))**2
  return ans
def estim_a(df):
  return (np.sqrt(sample_var(df)))

def estim_b(df):
  return aver_val(df, len(df)) - np.sqrt(sample_var(df))

omm_a = estim_a(df)
omm_b = estim_b(df)

print ("OMM: ", "\n", "estimate of the parameter а = ", omm_a,'\n', "estimate of the parameter b = ", omm_b)
```

```python
# ОМП
data = a1
x_min = min(data)

def aver_val(data, n):
  return np.sum(data)/n

def estim__a(data):
  return (aver_val(data, len(data)) - x_min)

def estim__b(data):
  return (x_min)

omp_a = estim__a(data)
omp_b = estim__b(data)

print ("OMП: ", "\n", "estimate of the parameter а = ", omp_a,'\n', "estimate of the parameter b = ", omp_b)
```
